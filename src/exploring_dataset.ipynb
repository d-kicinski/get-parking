```python
import pandas as pd
```

```python
pdf = pd.read_csv("dataset/parking_data.csv")
data = pdf['config_name']
pdf.columns
```

```python
pdf['geo'] = pdf[['latitude', 'longitude']].values.tolist()
#pdf = pdf[['geo','latitude','longitude', 'zone', 'area']]
pdf['geo'].values.tolist()
```

```python
pdf['area'].unique()
```

```python
time_max = data.str.extract('(?P<time_max>\w+\s\w+\s\w+)\s')
price = data.str.extract('\s(?P<price>\$\d+.\d+)\s')
hour_open = data.str.extract('\s(?P<hour_open>\d+[a-z]{2})-')
hour_close = data.str.extract('-(?P<hour_close>\d+[a-z]{2})\s')
day_open = data.str.extract('\s(?P<day_open>[A-Z][a-z]+)-')
day_close = data.str.extract('-(?P<day_close>[A-Z][a-z]+)')
pdf = pd.concat([pdf, time_max, price, hour_open, hour_close, day_open, day_close], axis=1)
```

```python
print("Prices: ", pdf["price"].unique())
print("time_max: ", pdf["time_max"].unique())
print("hour_open: ", pdf["hour_open"].unique())
print("hour_close: ", pdf["hour_close"].unique())
print("day_open: ", pdf["day_open"].unique())
print("day_close: ", pdf["day_close"].unique())
```

```python
print(pdf.columns)
print("latitude = ", max(pdf['latitude'].values[pdf['latitude'].values!=33]),min(pdf['latitude'].values[pdf['latitude'].values!=0]))
print("longitude = ", max(pdf['longitude'].values[pdf['longitude'].values!=0]),min(pdf['longitude'].values[pdf['longitude'].values!=-180]))
```
